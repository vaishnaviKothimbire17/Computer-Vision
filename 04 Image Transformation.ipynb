# Image Translation

```python
import cv2
import numpy as np

image = cv2.imread('lenna.png')

#store the ht & wth of the image
height, width = image.shape[:2]

quarter_height, quarter_width = height / 6, width / 4

T = np.float32([[1, 0, quarter_width], [0,1,quarter_height]])

#we use wrapapline to transform the image using Matrix
img_translation = cv2.warpAffine(image, T, (width,height))


# show the image with the ellipse
cv2.imshow('Originalimage',image)
cv2.imshow('Translation',img_translation)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

image = cv2.imread('lenna.png')

T = np.float32([[1, 0, -50], [0,0,-60]])

#we use wrapapline to transform the image using Matrix
img_translation = cv2.warpAffine(image, T, (width,height))


# show the image with the ellipse
cv2.imshow('Originalimage',image)
cv2.imshow('Translation',img_translation)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Image Rotation

```python
import cv2

# load the image
image = cv2.imread('lenna.png')

# Rotate the image 45 deg clockwise
rotated_image = cv2.rotate(image,cv2.ROTATE_180)

# show the rotated image
cv2.imshow('Rotated Image',rotated_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Image Rescaling

```python
import cv2

# load the image
image = cv2.imread('lenna.png')

# Resize the image with fixed aspect ratio
resized_image = cv2.resize(image,(200,300))

cv2.imshow('Resized Image',resized_image)

# show the rotated image
cv2.imshow('Resized Image',resized_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2

# load the image
image = cv2.imread('lenna.png')

cv2.imshow('Resized Image',resized_image)

# Resize the image with aspect ratio preserved
resized_image = cv2.resize(image,None, fx=1.25,fy=0.7)

# show the rotated image
cv2.imshow('Resized Image',resized_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## using Scikit Image

```python
from skimage.transform import resize

# load the image
image = cv2.imread('lenna.png')

# Resize the image with fixed aspect ratio
resized_image = resize(image,(900,900), anti_aliasing=True)

# Resize the image with aspect ratio preserved
resized_image = resize(image,(1000,1000), anti_aliasing=True, order=3)

# show the rotated image
cv2.imshow('Resized Image',resized_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Image Flipping

```python
import cv2

# load the image
image = cv2.imread('lenna.png')

# Flip the image horizontally
flipped_horizontal = cv2.flip(image,0)

# Flip the image vertically
flipped_vertical = cv2.flip(image,1)

# Flip the image diagonally
flipped_diagonal = cv2.flip(image,-1)

# show the rotated image
cv2.imshow('Original',image)
cv2.imshow('horizontal Image',flipped_horizontal)
cv2.imshow('vertical Image',flipped_vertical)
cv2.imshow('diagonal Image',flipped_diagonal)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Image Cropping

```python
import cv2

# load the image
image = cv2.imread('lenna.png')

new_img = image[200:300,200:300]
# show the rotated image
cv2.imshow('Cropped Image',new_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2

# load the image
image = cv2.imread('lenna.png')

height, width, channels = image.shape

startRow = int(height*.35)
startCol = int(width*.35)
endRow = int(height*.65)
endCol = int(width*.65)

new_img = image[startRow:endRow,startCol:endCol]

cv2.imshow('Original Image',image)
cv2.imshow('Cropped Image',new_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
