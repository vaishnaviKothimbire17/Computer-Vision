## Bitwise Operations

```python
import cv2
import numpy as np

image1 = cv2.imread('input1.png')
image2 = cv2.imread('input2.png')

cv2.imshow('input1',image1)
cv2.imshow('input2',image2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Image Masking

```python
import cv2
import numpy as np
```

```python
img = cv2.imread('rgy.jpg')

cv2.imshow('My Image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# the kernel to be used for dilation purpose
kernel = np.ones((5,5),np.uint8)

# converting the image to HSV format
hsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)

# defining the lower and upper va;ues of HSv, this will detect yellow color
Lower_hsv = np.array([20,70,100])
Upper_hsv = np.array([30,255,255])

# creating the mask buby eroding,morphing,dilating process
Mask = cv2.inRange(hsv,Lower_hsv,Upper_hsv)
Mask = cv2.erode(Mask,kernel,iterations = 1)
Mask = cv2.morphologyEx(Mask,cv2.MORPH_OPEN,kernel)
Mask = cv2.dilate(Mask,kernel,iterations = 1)

# inverting the mask by performing bitwisw not operations
Mask = cv2.bitwise_not(Mask)

# Displaying the image
cv2.imshow('Mask',Mask)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
mask_yellow = cv2.bitwise_not(Mask)
Mask = cv2.bitwise_and(img,img,mask = mask_yellow)
```

```python
cv2.imshow('Mask',Mask)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
