## 3D Line Drawing

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

# define start and end points of the line in 3d spaces
start_point = (10,22,33)
end_point = (44,52,67)

# create a plot
fig = plt.figure()
ax = fig.add_subplot(111,projection = '3d')

#plot the line
ax.plot([start_point[0],end_point[0]],
       [start_point[1],end_point[1]],
       [start_point[2],end_point[2]],'r-')
ax.plot([10,40],[55,35],[35,30],'g')

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Line')

# show the plot
plt.show()
```

## Rectangle using matplotlib

```python
import matplotlib.pyplot as plt
```

```python
# Define rectangle coordinates
x_min,y_min,x_max,y_max = 1,2,4,5

# create the rectangle patch
rectangle = plt.Rectangle((x_min,y_min),x_max - x_min,y_max - y_min,
                         color = 'red', alpha = 0.4)

# Add the rectangle to a plot and customize labels / title

plt.plot([x_min, x_max],[y_min,y_min],'b-',label = 'X_axis')
plt.plot([x_min, x_min],[y_min,y_max],'b-',label = 'Y_axis')
plt.gca().add_patch(rectangle)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Rectangle plot")
plt.legend()

#show the plot
plt.show()
```

## Ellipse using Matplotlib

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse

# define the center coordinates, radii & angle
center = (0,0)
width = 8
height = 10
angle = 60

# create the ellipse object 
ellipse = Ellipse(xy=center, width=width, height=height, angle=angle, color='blue')

# add the ellipse to the plot
fig, ax = plt.subplots()
ax.add_patch(ellipse)
ax.set_aspect('equal')

# set limits slightly wider than ellipse
ax.set_xlim(-width - width/5, width + width/5)
ax.set_ylim(-height - height/5,height + height/5)

# Add labels & title
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Ellipse plot")

#show the plot
plt.show()
```

# Drawing Circle

```python
import cv2

# load your image
image = cv2.imread('lenna.png')

# Define center coordinates, radii, & angle
center = (255,255) # (x,y)


# draw a blue circle with thk 5 pixels
for i in range(10,301,5):
    cv2.circle(image, center, i, (i//2,0,i),1)

# show the image with the ellipse
cv2.imshow('Image with Circle',image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt


# define the center coordinates, radii & angle
center = (0,0)
radius = 8

# create the circle object 
circle = plt.Circle(xy=center, radius = radius ,color='lightcoral')

# add the circle to the plot
fig, ax = plt.subplots()
ax.add_patch(circle)
ax.set_aspect('equal')

# set limits slightly wider than circle
ax.set_xlim(-radius - radius/5, radius + radius/5)
ax.set_ylim(-radius - radius/5,radius + radius/5)

# Add labels & title
plt.xlabel("X")
plt.ylabel("Y")
plt.title("circle plot")

#show the plot
plt.show()
```

# Event Handling

```python

top = Tk()
top.geometry('800x500')
top.title('My Window')

def show():
   Label(top, text="Your name:"+str(e.get()).upper()).place(x=50,y=150)

Label(top, text='Enter Your name:').place(x=50,y=50) # change position according your requirement
e = Entry(top)
e.place(x=50,y=80)
Button(top, text = 'Submit', command = show).place(x=50,y=110)
top.mainloop()
```

```python
root = Tk()
root.geometry('1000x800')
root.title('My Window')

# Create the canvas widget
canvas = Canvas(root, width=1000, height=800, bg="yellow")
canvas.pack()



def draw_line(event):
    x1,y1 = event.x, event.y
    canvas.create_line(x1,y1,x1+2,y1+2,width=2)

def draw_circle(event):
    x1,y1 = event.x, event.y
    canvas.create_oval(x1-10,y1-10,x1+10,y1+10,width=2, fill="black")

#Bind mouse event to drawing functions
canvas.bind("<Button-3>",draw_circle)
canvas.bind("<B1-Motion>",draw_line)
   
top.mainloop()
```

```python
# load your image
img = cv2.imread('Screenshot 2024-06-01 113224.png')

def draw_circle(event,x,y,flags,param):
   if event == cv2.EVENT_LBUTTONDOWN:
       cv2.circle(img,(x,y),5,(0,0,0),-1)

# Bind the callback function
cv2.namedWindow('image')
cv2.setMouseCallback('image',draw_circle)

# show the imag eand wait for events
while True:
    cv2.imshow('image',img)
    if cv2.waitKey(1) == ord('q'):
        break

cv2.destroyAllWindows()
```

```python
text ='lenna'
font_face = cv2.FONT_HERSHEY_SCRIPT_SIMPLEX
font_scale = 1.0
color =(0,255,145)
thickness = 2

# calculate text width & height
text_size, baseline = cv2.getTextSize(text, font_face, font_scale, thickness)

# define text & rectangle position
text_origin =(200,200)
box_origin =(text_origin[0] - 10, text_origin[0] - baseline - 20)
box_width, box_height = text_size[0] + 20 , text_size[1]+ 20

image = cv2.imread("lenna.png")

# draw the rectangle and text
cv2.rectangle(image, box_origin,(box_origin[0] + box_width,
                                box_origin[1] + box_height), color , thickness)
cv2.putText(image,  text, text_origin, font_face,font_scale,color, thickness)


# show the image with the ellipse
cv2.imshow('Image',image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
